# Example of INL calculation

```python
%matplotlib widget
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
```

create variable to select the reference source

```python
# source = Fluke 5720A
# dutb = HP3458A
ref = "source"
```

load Transmille measurement from xdevs server

it doesn't matter if the source is your local file system or a web address

```python
data = pd.read_csv("https://xdevs.com/doc/Transmille/8104/3458_vs_8104/dcl_10vdc_5720_8104r_3458a_nlpc100_raw_jul2021.csv", sep=";")
data
```

There are multiple measurements for each voltage point. Therefore, we group them and calculate the mean for every group

```python
data = data.groupby(['source']).apply(np.mean)
```

Now we can calculate a least square fit

```python
slope, intercept, rvalue, pvalue, stderr = stats.linregress(data["duta"], data[ref])
```

```python
corr  = data["duta"] * slope - intercept
error = corr - data[ref]

plt.figure()
plt.title("Transmille INL")
plt.plot(data[ref], error, '.', label="measured error")

plt.ylabel("Deviation in V")
plt.xlabel("Input in V")
plt.grid()
```
